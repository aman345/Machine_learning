```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
apple_data = pd.read_csv("AAPL.csv")
```

```python
apple_data
```

```python
apple_training_processed = apple_data.iloc[:,1:2].values
```

```python
apple_training_processed
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range =(0,1))
apple_training_scaled = scaler.fit_transform(apple_training_processed)
```

```python
features_set = []
labels = []
for i in range(60,250):
    features_set.append(apple_training_scaled[i-60:i,0])
    labels.append(apple_training_scaled[i,0])

features_set,labels = np.array(features_set),np.array(labels)
features_set = np.reshape(features_set,(features_set.shape[0],features_set.shape[1],1))
```

```python
#model 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences =True,input_shape=(features_set.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences =True))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences =True))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))
```

```python
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss = 'mean_squared_error')
```

```python
model.fit(features_set,labels,epochs=100,batch_size=32)
```

```python
apple_testing_complete =pd.read_csv("AAPL.csv")
apple_testing_processed = apple_testing_complete.iloc[:,1:2].values
```

```python
apple_total = pd.concat((apple_data['Open'],apple_testing_complete['Open']),axis=0)
```

```python
test_inputs = apple_total[len(apple_total)-len(apple_testing_complete)-60:].values
```

```python
test_inputs = test_inputs.reshape(-1,1)
test_inputs = scaler.transform(test_inputs)
```

```python
test_features = []
for i in range(60,80):
    test_features.append(test_inputs[i-60:i,0])
```

```python
test_features = np.array(test_features)
test_features = np.reshape(test_features,(test_features.shape[0],test_features.shape[1],1))
```

```python
predictions = model.predict(test_features)
```

```python
predictions = scaler.inverse_transform(predictions)
```

```python
plt.figure(figsize=(10,6))
plt.plot(apple_testing_processed,color='blue',label='Actual prediction')
plt.plot(predictions,color='red',label='Predicted stock value')
plt.title("Stock Prediction ")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
